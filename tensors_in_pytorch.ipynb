```python
import time
size=10000
maxtrix1=torch.rand(size,size)
maxtrix2=torch.rand(size,size)

#Measure time on cpu
start_time=time.time()
result_cpu=torch.matmul(maxtrix1,maxtrix2)
end_time=time.time()
print("Time taken on cpu",end_time-start_time)

mat1_gpu=maxtrix1.to('cuda')
mat2_gpu=maxtrix2.to('cuda')
#Measure time on gpu
start_time=time.time()
result_gpu=torch.matmul(mat1_gpu,mat2_gpu)
end_time=time.time()
print("Time taken on gpu",end_time-start_time)
#
```
